```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
```

```python
# baca file csv
data_jakarta = pd.read_csv('/home/rizal/Documents/PUNYA_AQ/DATA-DKI-MENURUT-PENDIDIKAN-TAHUN-2014-ver-1-xlsx.csv')
```

```python
# print sample data jakarta
data_jakarta.head()
```

```python
# menampilkan jumlah data masing-masing kota/kabupaten
data_jakarta['nama_kabupaten/kota'].value_counts()
```

```python
kabupaten_kota = data_jakarta['nama_kabupaten/kota'].unique()
for i in kabupaten_kota:
    print(i)
```

```python
data_pendidikan_jakarta = []
for i in kabupaten_kota:
    tmp = data_jakarta['nama_kabupaten/kota'] == i
    data_pendidikan_jakarta.append((i,data_jakarta[tmp][['tidak/belum_sekolah', 'belum_tamat_sd/sederajat', 'tamat_sd/sederajat', 
                                'sltp/sederajat', 'slta/sederajat', 'diploma_I/II', 'akademi/diploma_III/s._muda',
                                'diploma_IV/strata I', "strata_II", 'strata_III']].sum()))
```

```python
for i,j in data_pendidikan_jakarta:
    j.plot(kind='barh', title=i, figsize=(15, 5))
    plt.show()
```
